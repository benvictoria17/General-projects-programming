```python
import re

def operator(criteria):
    d = {}
    if ("<>" in criteria):
        d["operator"] = "<>"
    elif ("<=" in criteria):
        d["operator"] = "<="
    elif (">=" in criteria):
        d["operator"] = ">="
    elif (">" in criteria):
        d["operator"] = ">"
    elif ("<" in criteria):
        d["operator"] = "<"
    else:
        return -1
    
    try:
        d["num"] = float(re.search(r'[+-]?\d+\.\d+', criteria).group())
    except:
        d["num"] = int(re.search(r'\d+', criteria).group())
        
    return d

def count_if(values, criteria):
    count = 0
    if (type(criteria) == int or type(criteria) == float):
        for i in values:
            if (i == criteria):
                count += 1
    elif (type(criteria) == str and operator(criteria) == -1):
        for i in values:
            if (i == criteria):
                count += 1
    elif (operator(criteria) != -1):
        d = operator(criteria)
        for i in values:
            if (d["operator"] == "<>"):
                if (i != d["num"]):
                    count += 1
            elif (d["operator"] == "<="):
                if (i <= d["num"]):
                    count += 1
            elif (d["operator"] == ">="):
                if (i >= d["num"]):
                    count += 1
            elif (d["operator"] == ">"):
                if (i > d["num"]):
                    count += 1
            elif (d["operator"] == "<"):
                if (i < d["num"]):
                    count += 1
    return count
    
def sum_if(values, criteria):
    newList = []
    if (type(criteria) == int or type(criteria) == float):
        for i in values:
            if (i == criteria):
                newList.append(i)
    else:
        d = operator(criteria)
        for i in values:
            if (d["operator"] == "<>"):
                if (i != d["num"]):
                    newList.append(i)
            elif (d["operator"] == "<="):
                if (i <= d["num"]):
                    newList.append(i)
            elif (d["operator"] == ">="):
                if (i >= d["num"]):
                    newList.append(i)
            elif (d["operator"] == ">"):
                if (i > d["num"]):
                    newList.append(i)
            elif (d["operator"] == "<"):
                if (i < d["num"]):
                    newList.append(i)
    
    fsum = sum(newList)
    return fsum
    
def average_if(values, criteria):
    newList = []
    if (type(criteria) == int or type(criteria) == float):
        for i in values:
            if (i == criteria):
                newList.append(i)
    else:
        d = operator(criteria)
        for i in values:
            if (d["operator"] == "<>"):
                if (i != d["num"]):
                    newList.append(i)
            elif (d["operator"] == "<="):
                if (i <= d["num"]):
                    newList.append(i)
            elif (d["operator"] == ">="):
                if (i >= d["num"]):
                    newList.append(i)
            elif (d["operator"] == ">"):
                if (i > d["num"]):
                    newList.append(i)
            elif (d["operator"] == "<"):
                if (i < d["num"]):
                    newList.append(i)
    
    avg = sum(newList) / float(len(newList))
    return avg
```
